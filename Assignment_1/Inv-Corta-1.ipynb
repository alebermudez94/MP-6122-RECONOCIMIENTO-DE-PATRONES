# Tutorial de limpieza de datos

Para empezar algo muy importante es saber si en nuestro DataFrame faltan valores, esto resulta muy facil con Pandas usando **pd.isnull**
Esta funcion devuelve una matriz de true/false
True: Valores faltantes
False: Valores no faltantes

```python
#Comprobar si hay datos nulos en el DataFrame
data_df.isnull()
```

Para manejar mejor el DataFrame se puede sumar valores faltantes por categoria, esto se logra con la funcion **pd.sum()** de la siguiente manera:

```python
#Suma de datos faltantes en el DataFrame
data_df.isnull().sum()
```

Para eliminar estos datos perdidos se hace uso de:
1. **data_df.dropna()** Para eliminar filas
2. **data_df.dropna(axis = 'columns')** Para eliminar columnas

```python
#Borrar las filas donde al menos un elemento es faltante
data_df.dropna()
```

```python
#Borrar las columnas donde al menos un elemento es faltante
data_df.dropna(axis='columns')
```

```python
#Borrar las filas donde todos los elementos son faltantes
data_df.dropna(how='all')
```

```python
#Borrar las filas donde al menos 2 elementos son faltantes
data_df.dropna(thresh=2)
```

```python
#Definir cuales columnas buscar por valores faltantes
data_df.dropna(subset=['Mes','peso'])
```

En caso contrario de que no se quiera eliminar datos, sino rellanar los valores perdidos con otros valores se hace uso de la siguiente instruccion:
    data_df.fillna(w)
Estos valores perdidos se reemplazan por la variable **w**, pero puede tomar 
1. **Valores numerico**
2. **La media de los datos**
3. **Letras**.

Los ejemplos se muestran de la siguiente manera:

```python
#Reemplaza los valores perdidos por la media
data_df.fillna(data_df.mean ())
```

```python
#Reemplaza los valores perdidos por la media
data_df.fillna('x')
```

# Exploratory Data Analysis (EDA)

```python
# Lo primero que hay que hacer es cargar el dataset

# Hay 2 formas de hacer esto, la primera es descargar la base de datos, encontrar una ruta y luego cargarla en el notebook:

## Definiendo el directorio

#os.chdir("C:\Users\bermudal\Reconocimiento de patrones")
#dataset= pd.read_csv('iris.csv')

# La segunda forma es llamar a una biblioteca que ya tiene el conjunto de datos:

from sklearn.datasets import load_iris

iris_data=load_iris()
dataset=pd.DataFrame(data= np.c_[iris_data['data'], iris_data['target']],
                     columns= iris_data['feature_names'] + ['target'])

# En este caso vamos a utilizar el archivo provisto en la biblioteca.
```

```python
# Veamos la cabecera del conjunto de datos.

dataset.head()
```

```python
# En este caso, la variable de destino es básicamente una clasificación codificada, veamos qué otros valores tienen destino

dataset['target'].unique()
```

```python
# Parece que está compuesto por tres valores "0,1,2", convirtamos estos valores en algún texto.
# Pero primero, veamos un resumen del conjunto de datos, la función de descripción nos permitirá hacer eso

dataset.describe()
```

```python
# La funcion "describe" nos permitirá tener algunas estadísticas del conjunto de datos, en este caso, la variable de destino no necesita
# este tipo de estadísticas ya que es una variable categórica codificada
# Ahora, veamos si faltan algunos valores en las columnas

dataset.isnull().sum()
```

```python
# Parece que no faltan valores, en caso de que necesite limpiarlos, por favor repasar el tutorial anterior o mire este código:

# Primero puedes eliminar filas de valores na con dataset.dropna ()

# O puede llenar valores na con cualquier otro valor: datast.fillna (0)

# Cambiemos la codificación de destino con los nombres "A", "B", "C" para realizar algunos gráficos
dataset['target']=dataset['target'].replace({0: 'A', 1: 'B', 2: 'C'}) #Aquí estoy reemplazando

# Veamos el dataframe de nuevo
dataset['target'].unique()
```

```python
# Ahora puede ver cómo "0", "1", "2" se cambió a "A", "B", "C"

# Hagamos un promedio de la longitud del sépalo para entender si hay algún tipo de planta que tenga más longitud que las otras
x = dataset.groupby('target')['sepal length (cm)'].mean().sort_values()

x
```

```python
# Parece que "C" es más grande que los otros

# Vamos a trazar eso

sepal_mean_plot=x.plot(kind='barh')

sepal_mean_plot.set_xlabel("Sepal mean cm", labelpad=20, weight='bold', size=12)
```
